```python
""" ChE421: HW 5 Code Portion (12/15/2018):

The loops for method of lines are defined as separate functions. There are two functions (ODE_ox, ODE_T) which 
take in vectors for C and T to output vectors dCdz and dTdz. 

The main function MOL takes an inlet temperature and wall temperatue and performs method of lines by solving 
the system of ODEs along z. It will create matrices to store C and T for each node at each z. Method for 
solving the ODE system is Euler.

The MOL function is currently already called to create plots similar to Figures 9.3 and 9.4. When rho = 1.24
and ep = 1.0, plots were found to match the textbook's plots.

The code also creates a plot of pressure which was assumed to be uniform across nodes. Question 6 answers are
in a section towards the end. """

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

### Define Parameters

```python
#  General Parameters
ep = 1.0 # Void fraction       
v_s = 1.0 # [m/s] Superficial velocity
d_t = 0.05 #[m] Inner diameter
radius = d_t/2.0 # [m] Inner radius 
L = 5.0 #[m] Length of tube
k_0 = np.exp(20.348) #[1/s] Parameter from equation for rate constant             
T_act = 13636 #[K] Activation Temperature
R_g = 8.314 #(m^3-atm/mol-K)

# Concentration ODE Parameters
D_r = 3.8e-4 # [m^2/s]  Radial Diffusivity          
c_ox_in = 0.4144  #[mol/m^3] Inlet concentration

# Temperature ODE Parameters
delH  = -307000.0 #[cal/mol]  Enthalpy of reaction
Twall = 640.0   # [K] Wall temperature
kappa = 0.4   #[cal/m-s-K] Thermal conductivity
Cp = 0.237 # [cal/g*K] Heat capacity 
heat = -307000.0 # [cal/mol] heat of reaction to mass units
rho = 1240.0 #[g/m^3] Gas density at the inlet             
hr = 120.0  #[cal/s-m^2-K] heat transfer coefficient
T_in = 300.0 # [K]  Inlet temperature            
E_r = 1.3e-3 #[m^2/s] Effective thermal diffusivity

# Method of Lines Parameters
N = 10 # Number of nodes
r = np.linspace(0, radius, N) # Create node positions along radius
h = r[1] - r[0]   # Radial step size           
zstep = v_s*h**2 / 4 / E_r # Axial step size (Equation 9.8)
z = np.arange(0,L,zstep)   # Create points along z

# Pressure Parameters
P_in = 2.2*101325 #[Pa] Pressure (assuming no pressure drop)
mu = 3.0e-5  # [Pa s] Viscosity of mixture
d_p = 0.003 #[m] Particle diameter
Re_p = rho*1e-3*d_p*v_s/mu 
```

### Define functions to find solutions for varying conditions

```python
# Creates dCdz vector
def ODE_ox(a, T):
        dCdz = np.zeros(N)  # Initialize derivatives for each node
        dCdz[0] = (4*D_r/(v_s*h**2))*(a[1]-a[0]) - (ep*k_0/v_s)*np.exp(-T_act/T[0])*a[0]  # Center-line boundary condition
        for i in range(1,N-1):
                C_p = D_r*(1+2*i)/(h**2 * 2*i) # C+ from table 8.2 in textbook
                C_m = D_r*(-1+2*i)/(h**2 * 2*i) # C0 from table 8.2 in textbook
                C_0 = -2*D_r/(h**2)             # C- from table 8.2 in textbook
                dCdz[i] = (C_p*a[i+1] + C_0*a[i] + C_m*a[i-1] - ep*k_0*np.exp(-T_act/T[i])*a[i])/v_s # Equation 8.31
        return dCdz

# Creates dTdz vector
def ODE_T(C, T):
        dTdz = np.zeros(N) # Initialize derivatives for each node
        dTdz[0] = (4*E_r/(v_s*h**2))*(T[1]-T[0]) - (ep*k_0*delH*C[0]/(v_s*rho*Cp))*np.exp(-T_act/T[0])# Center-line boundary condition
        for i in range(1,N-1):
                E_p = kappa*(1+2*i)/(h**2 * 2*i)  # E+ from table 8.3 in textbook
                E_m = kappa*(-1+2*i)/(h**2 * 2*i) # E0 from table 8.3 in textbook
                E_0 = -2*kappa/(h**2)             # E- from table 8.3 in textbook
                dTdz[i] = (E_p*T[i+1] + E_0*T[i] + E_m*T[i-1] - ep*k_0*delH*C[i]*np.exp(-T_act/T[i]))/(rho*Cp*v_s) # Equation 8.33
        return dTdz
        
# Solves system of ODEs along z using Euler's method
def MOL(T_inlet,T_wall): 
        ox = c_ox_in*np.ones(N) # Becomes initial value in loop
        T = T_inlet*np.ones(N)   # Initial temperature in nodes

        ox_mat = np.zeros((len(ox),len(z))) # Store concentration of nodes in a matrix
        T_mat = np.zeros((len(T), len(z)))  # Store temperature of nodes in a matrix
        P = np.zeros(len(z))

        # Initialize matrices with vectors of initial conditions
        ox_mat[:,0] = ox
        T_mat[:,0] = T 
        P[0] = P_in
        p = P_in

        # Euler Loop
        for i in range(len(z)-1):
                ox_change = ODE_ox(ox,T) # Calculate change in OX using function to find dCdz
                T_change = ODE_T(ox,T)   # Calculate change in T using function to find dTdz
                dPdz = -1763.89 # Pressure change

                # Update Loop Values
                ox += zstep*ox_change
                T += zstep*T_change
                p += zstep*dPdz
  
                # Apply Boundary Conditions at the Wall
                ox[N-1] = (18.0*ox[N-2] - 9.0*ox[N-3] + 2.0*ox[N-4])/11.0   # Equation 8.32
                T[N-1] = (6*h*hr*T_wall + kappa*(18*T[N-2] - 9*T[N-3] + 2*T[N-4]))/(6*h*hr + 11*kappa) # Equation 9.10

                # Update Matrices
                ox_mat[:,i+1] = ox
                T_mat[:,i+1] = T
                P[i+1] = p
        return ox_mat, T_mat, P             # Return both matrices and P vector


#Plot based on Figure 9.3 (Max centerline temperature based on T_wall)

# Create vector of wall temperatures from 630 to 650 K (used for function and plotting)
T_walls = [630+i for i in range(21)]

# Make a function to plot the two graphs of maximum centerline temperature
def wall_plot(T_inlet):

        T_wall_vec = np.array(T_walls)                    # Import list of T_wall values to function
        T_max_vec = np.zeros(T_wall_vec.shape)            # Create vector to store max T along centerline

        for i in range(len(T_wall_vec)):
                T_wall = T_wall_vec[i]               # Set wall T
                T_matr = MOL(T_inlet,T_wall)[1]      # Calculate T matrix
                max_T = np.amax(T_matr[0,:])         # Calculate maximum along centerline 
                T_max_vec[i] = max_T                 # Store max centerline T for particular T wall

        return T_max_vec                # Returns vector of maximum temperatures at each T_wall
```

#### Solve problems and plot solutions

```python
plt.figure(1)
# Plot max temperatures from function against wall temperatures
plt.plot(T_walls, wall_plot(600), label='T$_{in}$ = 600K') # T_in = 600K
plt.plot(T_walls, wall_plot(300), label='T$_{in}$ = 300K') # T_in = 300K
plt.xlabel('Wall Temperature, K')
plt.ylabel('Maximum Centerline Temperature, K')
plt.title('Maximum Centerline Temperature for Inlet Temperatures of 300K and 600K')
plt.legend(loc='best')
plt.ylim(500, 1000) # Use same limits as Fig 9.3


# Plot based on Figure 9.4 (Centerline Temperature vs. Axial Position)
ox_300, T_300, P_300 = MOL(300,Twall) # Return matrices of values for T_in = 300K, Twall = 640K
ox_600, T_600, P_600 = MOL(600,Twall) # Return matrices of values for T_in = 600K, Twall = 640K

# Create dimensionless axial position
zeta = [z[i]/L for i in range(len(z))] # Divide each z[i] by L to get dimensionless position


plt.figure(2)
# Plot centerline temperatures against z. 
plt.plot(zeta, T_300[0,:], label='T$_{in}$ = 300K') # T_in = 300K (Look at all zeta for r = 0 in T_300)
plt.plot(zeta, T_600[0,:], label='T$_{in}$ = 600K') # T_in = 600K (Look at all zeta for r = 0 in T_600)
plt.xlabel('Dimensionless Axial Position')
plt.ylabel('Centerline Temperature, K')
plt.title('Plot of Centerline Temperature for Inlet Temperatures of 300K and 600K')
plt.legend(loc='best')
plt.ylim(300, 800) # Use same limits as Fig 9.4

plt.show()


# Question 5: 9.2 
# Use subscript 5 for problem 5
Ox_5, T_5, Pr = MOL(600,640) # Retrieve Pr vector for T_in = 600K and T_wall= 640K

# Find pressure, OX concentration, and temperature in and out of reactor
Pr_in = Pr[0]
Pr_out = Pr[-1]
Ox_5_in = Ox_5[0,0]
Ox_5_out = Ox_5[0,-1]
T_5_in = T_5[0,0]
T_5_out = T_5[0,-1]


C_in =  Pr_in/R_g/T_5_in  # Concentration of gas in
C_air_in = C_in - Ox_5_in # Concentration of air in
C_o2_in = 0.21*C_air_in   # Concentration of O2 in

# Change in Pressure (Pa)
delta_P = (Pr[0]-Pr[-1]) 

# Change in OX concentration
delta_C_ox = Ox_5_in - Ox_5_out 

# Reaction is: OX + 1.5 O2 ==> PA
C_o2_out = C_o2_in - 1.5*delta_C_ox  # O2 is also consumed by reaction (stoichiometry)


Ratio_inlet = C_o2_in/Ox_5_in   # Oxygen to orthoxylene ratio at the inlet 
Ratio_outlet = C_o2_out/Ox_5_out # Oxygen to orthoxylene ratio at the outlet 


print("Question 5:")
print("The pressure drop is",round(delta_P), 'Pa which is small enough for the pseudohomogeneous assumption.')
print("The ratio at the outlet of C_O2 to C_OX is ", round(Ratio_outlet))
```
